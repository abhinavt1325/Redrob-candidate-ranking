```python
import pandas as pd
import json
df=pd.read_json("candidates.jsonl", lines=True)
```

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop_duplicates(subset='candidate_id')
print(df_clean['candidate_id'].duplicated().sum())
```

```python
df_clean.isnull().sum()
```

```python
signal_df = pd.json_normalize(df_clean['redrob_signals'])

signal_df.columns = ['signal_' + col for col in signal_df.columns]

df_clean = pd.concat(
    [df_clean.drop(columns=['redrob_signals']), signal_df],
    axis=1
)
```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
print(json.dumps(df_clean['profile'][0], indent=2))
```

```python
profile_df = pd.json_normalize(df_clean['profile'])

profile_df.columns = [
    'profile_' + col
    for col in profile_df.columns
]
df_clean = pd.concat(
    [df_clean.drop(columns=['profile']),
     profile_df],
    axis=1
)
```

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
print(type(df_clean['career_history'][0]))
print(len(df_clean['career_history'][0]))
print(json.dumps(df_clean['career_history'][0][0], indent=2))
```
