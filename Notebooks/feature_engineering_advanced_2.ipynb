```python
import pandas as pd
import json
df_feat=pd.read_pickle("dataset_features_v2.pkl")
```

```python
df_feat['skills_text'] = df_feat['skills'].apply(
    lambda x: ' '.join(skill['name'] for skill in x)
)
```

```python
df_feat['career_text'] = df_feat['career_history'].apply(
    lambda x: ' '.join(
        f"{job['title']} {job['industry']} {job['description']}"
        for job in x
    )
)
```

```python
df_feat['education_text'] = df_feat['education'].apply(
    lambda x: ' '.join(
        ed['degree'] + ' ' + ed['field_of_study']
        for ed in x
    )
)
```

```python
df_feat['profile_text'] = (
    df_feat['profile_headline'].fillna('') + ' ' +
    df_feat['profile_summary'].fillna('')
)
```

```python
df_feat['candidate_text'] = (
    df_feat['profile_text'] + ' ' +
    df_feat['skills_text'] + ' ' +
    df_feat['career_text'] + ' ' +
    df_feat['education_text']
)
```
