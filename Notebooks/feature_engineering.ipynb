```python
import pandas as pd
import json
df_clean = pd.read_pickle("dataset_cleaned.pkl")
```

```python
df_feat=df_clean.copy()
```

```python
df_feat['num_skills'] = df_feat['skills'].apply(len)
df_feat['num_skills'].describe()
```

```python
df_feat['num_jobs'] = df_feat['career_history'].apply(len)
df_feat['num_jobs'].describe()
```

```python
df_feat['num_certifications'] = df_feat['certifications'].apply(len)
df_feat['num_certifications'].describe()
```

```python
df_feat['num_languages'] = df_feat['languages'].apply(len)
df_feat['num_languages'].describe()
```

```python
df_feat['num_degrees'] = df_feat['education'].apply(len)
df_feat['num_degrees'].describe()
```

```python
df_feat['signal_open_to_work_flag'] = (
    df_feat['signal_open_to_work_flag']
    .astype(int)
)
```

```python
df_feat['signal_willing_to_relocate'] = (
    df_feat['signal_willing_to_relocate']
    .astype(int)
)
```

```python
verification_cols = [
    'signal_verified_email',
    'signal_verified_phone',
    'signal_linkedin_connected'
]
for col in verification_cols:
    df_feat[col] = df_feat[col].astype(int)
```

```python
df_feat['signal_last_active_date'].dtype
```

```python
df_feat['signal_last_active_date'] = pd.to_datetime(
    df_feat['signal_last_active_date'],
    errors='coerce'
)

df_feat['signal_signup_date'] = pd.to_datetime(
    df_feat['signal_signup_date'],
    errors='coerce'
)
```

```python
df_feat['signal_last_active_date'].dtype
df_feat['signal_signup_date'].dtype
```

```python
today = pd.Timestamp.today()

df_feat['days_since_last_active'] = (
    today - df_feat['signal_last_active_date']
).dt.days
```

```python
df_feat['days_since_signup'] = (
    today - df_feat['signal_signup_date']
).dt.days
```

```python
df_feat.to_pickle("dataset_features_v1.pkl")
```
